## Question
Calculate the bending moment and shear force of a simply supported beam of length l(m) carrying a uniformly distributed load of q (KN/m)

```python
# Asking users for Input parameters
#l = input("What is the length of the beam in m? ")
#l = float(l) #m
#q = input ("What s the magnitude of the load in KN/m? ")
#q = float (q) #KN/m
```

```python
l = 5 #m
q = 20 #KN/m
```

# Calculating Bending moment and Shear force


           Reactions and co-ordinate system

## Equilibrium equations 
$H_A = 0$  

$V_A + V_B -\frac{ql}{2} = 0$  

$V_Bl - \frac{ql^2}{2} = 0$

By solving the above equations one obtains the following  

$H_A = 0$ 

$V_A = V_B = \frac{ql}{2}$

At a point **x** along the length of the beam, one obtains the Bending moment **M** and the Shear force **V** :

$M_x = V_Ax - \frac{qx^2}{2} = \frac{q}{2}(lx - x^2)$

$V_x = V_A - qx = q ( \frac{l}{2} - x)$

$M_x$ and $V_x$ is now coded with python

```python
#x represents points along the beam length with M and V are calculated
x = np.linspace (0,l,30) #vector (30 elements) of evenly spaced numbers from 0 to l, 
M_x = q/2 *(l*x - x**2) #M equation
V_x = q * (l/2 -x) #V equation
```

## Ploting Bending moment and shear force diagram 

```python

plt.figure(figsize=(10,4))
plt.plot(x,-M_x,label='Moment')
plt.plot(x,V_x, label = "Shear Force")
plt.title("Bending moment and shear force diagram", fontsize=15)
plt.xlabel('x', fontsize=15)
plt.ylabel('Moment or Shear Force', fontsize=15)
plt.legend()
plt.axhline(y=0, color='k')
#plt.axvline(x=0, color='k')
```
